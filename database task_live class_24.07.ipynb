```python
#3. read these dataset in pandas as a dataframe
import mysql.connector as conn
import pandas as pd
mydb=conn.connect(host="localhost",user="root", passwd="MySqlRoot!3456",use_pure=True)
cur=mydb.cursor()
cur.execute('use employees')
def select_column(d):
    cur=mydb.cursor()
    cur.execute(f'show columns from {d}')
    res=cur.fetchall()
    cols=[]
    for c in res:
        cols.append(c[0])
    return cols
cur.execute('select*from attr')
res1=cur.fetchall()
cur.execute('select*from dress_sales')
res2=cur.fetchall()
df_attr=pd.DataFrame(res1,columns=select_column('attr'))
df_dress=pd.DataFrame(res2,columns=select_column('dress_sales'))
```
