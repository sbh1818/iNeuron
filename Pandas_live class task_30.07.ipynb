```python
import pandas as pd
agent_performance=pd.read_csv('AgentPerformance (1).xlsx - Agent Performance Report.csv')
agent_login=pd.read_csv('Agent_Login_Report (4).xls - Agent Login Report.csv')
```

```python
#1 .Find out there avarage rating on weekly basis keep this in a mind that they take two days of leave
agent_performance['week']=agent_performance.Date.dt.week
```

```python
agent_performance.drop(agent_performance[agent_performance['Total Feedback']==0].index,axis=0,inplace=True) #to take care of weekly offs and 0 feedback counts
```

```python
agent_performance.groupby('week')['Average Rating'].mean()
#remember missing agents did not work for a full week
```

```python
#2 .Total working days for each agents
agent_performance.groupby('Agent Name').count()['Date']
```

```python
#3. Total query that you hvae taken
agent_performance.groupby('Agent Name').sum('Total Chats')['Total Chats']
```

```python
#4. total Feedback that you have received 
agent_performance.groupby('Agent Name').sum('Total Feedback')['Total Feedback']
```

```python
#5. agent name who have average rating between 3.5 to 4
agent_performance[(agent_performance['Average Rating']>=3.5) & (agent_performance['Average Rating']<=4)].groupby('Agent Name').first()['Average Rating']
```

```python
#6 . Agent name who have rating lesss then 3.5
agent_performance[agent_performance['Average Rating']<3.5].groupby('Agent Name').first()['Average Rating']
```

```python
#7 . agent name who have rating more then 4.5
agent_performance[agent_performance['Average Rating']>4.5].groupby('Agent Name').first()['Average Rating']
```

```python
#8 . how many feedaback agents have received more then 4.5 average
agent_performance[agent_performance['Average Rating']>4.5].groupby('Agent Name').first()['Total Feedback']
```

```python
#9 . average weekly response time for each agent
agent_performance.groupby('week').first()[['Agent Name','Average Response Time']]
#agent_performance['Average Response Time']=pd.to_datetime(agent_performance['Average Response Time']) to convert Average Response Time to datetime object
#remember missing agents did not work for a full week
```

```python
#10 . average weekely resolution time for each agents
#agent_performance['Average Resolution Time']=pd.to_datetime(agent_performance['Average Resolution Time'])# converting Average Resolution Time to datetime object
agent_performance.groupby('week').first()[['Agent Name','Average Resolution Time']]
#remember missing agents did not work for a full week
```

```python
#11 . list of all agents name
agent_performance.groupby('Agent Name').first().index
```

```python
#12. percentage of chat on which they have received a feedback 
a=agent_performance.groupby('Agent Name').sum('Total Chats')
a['feedback %']=a['Total Feedback']/a['Total Chats']*100
a['feedback %']
```

```python
#13 . Total contributation hour for each and every agents weekly basis 
#agent_login['week']=agent_login['Date'].dt.week to add a week column in the data frame
```

```python
agent_login['dt_duration']=agent_login['Duration'].apply(lambda x :int(x[:2])*3600+int(x[3:5])*60+int(x[6:]))
#agent_login['dt_duration']=pd.to_datetime(agent_login['dt_duration'],unit='s')
#had to do some work around as could not convert the 'Duration' column into date time straightaway
b=agent_login.groupby('Agent').sum('dt_duration')
b['dt_duration']=pd.to_datetime(b['dt_duration'],unit='s')
b['dt_duration']# couldn't care to remove '1970-01-01'
#remember all agents did not work for a full week
```

```python
#14. total percentage of active hour for a month
agent_login['month']=agent_login['Date'].dt.month
c=agent_login.groupby('Agent').first()
c['active hour %']=c['dt_duration']*100/8*3600
c
```
